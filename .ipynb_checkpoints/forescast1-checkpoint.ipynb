```python
import numpy as np
import pandas as pd
import seaborn as sns 
from matplotlib import pyplot as plt

from statsmodels.tsa.api import ExponentialSmoothing, SimpleExpSmoothing, Holt
from sklearn.linear_model import LinearRegression

import warnings
warnings.filterwarnings("ignore")
```

```python
df=pd.read_csv("gold_monthly_csv.csv")
#df.head() #show first 5 observations in database
#df.shape #show how many observations and variables the database has.
```

```python
df_1=df.groupby(df.index.year).mean().rename(columns={'Price':'Mean'})
df_1=df_1.merge(df.groupby(df.index.year).std().rename(columns={'Price': 'Std'}), left_index=True, right_index=True)
df_1['Cov_pct']=((df_1['Std']/df_1['Mean'])*100).round(2)
df_1.head()
```

```python
fig, ax=plt.subplots(figsize=(15,10))
df_1['Cov_pct'].plot();
plt.title('Average Gold price yearly since 1950')
plt.xlabel('Year')
plt.ylabel('CV in %')
plt.show();
```

```python
train=df[df.index.year<=2015]
test=df[df.index.year>2015]

print(train.shape)
print(test.shape)
```

```python
train['Price'].plot(figsize=(13,5), fontsize=15)
test['Price'].plot(figsize=(13,5), fontsize=15)
plt.legend(['Training Data', 'Test Data'])
plt.show()
```

```python
train_time=[i+1 for i in range(len(train))]
test_time=[i+len(train)+1 for i in range(len(test))]
len(train_time), len(test_time)
```

```python
LR_train=train.copy()
LR_test=test.copy()
```

```python
LR_train['Time']=train_time
LR_test['Time']=test_time
```

```python
lr=LinearRegression()
lr.fit(LR_train[['Time']], LR_train['Price'].values)
```

```python
test_predictions_model1=lr.predict(LR_test[["Time"]])
LR_test["forecast"]=test_predictions_model1

plt.figure(figsize=(14,6))
plt.plot(train['Price'], label='Train')
plt.plot(test['Price'], label='Test')
plt.plot(LR_test['forecast'], label='Regression on time_test data')
plt.legend(loc='best')
plt.grid();
```

```python
def mape(actual, pred):
    return round((np.mean(abs(actual-pred)/actual))*100, 2)
```

```python
mape_model1_test=mape(test['Price'].values,test_predictions_model1)
print("MAPE is %3.3f"%(mape_model1_test), "%")
```

```python
results=pd.DataFrame({'Test Mape (%)': [mape_model1_test]}, index=['Regression on Time'])
results
```

```python
Naive_train=train.copy()
Naive_test=test.copy()
```

```python
Naive_test['Naive']=np.asarray(train['Price'])[len(np.asarray(train['Price']))-1]
Naive_test['Naive'].head()
```

```python
plt.figure(figsize=(12,8))
plt.plot(Naive_train['Price'], label='Train')
plt.plot(test['Price'], label='Test')
plt.plot(Naive_test['Naive'], label='Naive Forecast on Test Data')
plt.legend(loc='best')
plt.title('Naive Forecast')
plt.grid();
```

```python
mape_model2_test=mape(test['Price'].values,Naive_test['Naive'].values)
print("for Naive Forecast on the Test Data, MAPE is %3.3f"%(mape_model2_test), "%")
```

```python
resultsDf_2=pd.DataFrame({'Test Mape (%)': [mape_model2_test]}, index=['Naive Model'])
results=pd.concat([results,resultsDf_2])
results
```

```python
final_model = ExponentialSmoothing(df,
                                  trend='additive',
                                  seasonal='additive').fit(smoothing_level = 0.4,
                                                          smoothing_trend = 0.3,
                                                          smoothing_seasonal = 0.6)
```

```python
MAPE_final_model=mape(df['Price'].values, final_model.fittedvalues)
print('MAPE: ', MAPE_final_model)
```

```python
predictions = final_model.forecast(steps=len(test))
```

```python
pred_df=pd.DataFrame({"lower_CI": predictions - 1.96*np.std(final_model.resid, ddof=1),
                   "prediction":  predictions,
                   "upper_CI":  predictions + 1.96*np.std(final_model.resid, ddof=1)})
pred_df.head()
```

```python
axis = df.plot(label = 'Actual', figsize=(16,9))
pred_df['prediction'].plot(ax=axis, label='Forecast', alpha=0.5)
axis.fill_between(pred_df.index, pred_df['lower_CI'], pred_df['upper_CI'], color='m', alpha=0.15)
axis.set_xlabel('year-month')
axis.set_ylabel('price')
plt.legend(loc='best')
plt.grid()
plt.show()
```
